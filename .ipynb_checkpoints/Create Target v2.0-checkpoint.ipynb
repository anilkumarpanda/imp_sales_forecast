```python
def getPhar(hour):
    if hour in ['5:00','6:00','7:00','8:00','9:00','10:00','11:00','12:00']:
        return "Morning"
    if hour in ['13:00','14:00','15:00','16:00']:
        return "Afternoon"
    if hour in ['17:00','18:00','19:00','20:00','21:00']:
        return "Evening"
    else :
        return "Night"

def merge_hr(df):
    df = df.fillna(0)
    temp_df = pd.DataFrame()
    for col in df.columns[1:-1]:
        #print (col.split(" "))
        if len(col.split(" "))>=2:
            phar = getPhar(col.split(" ")[1])
            if phar in temp_df.columns:
                temp_df[phar] = temp_df[phar]+df[col]
            else :
                temp_df[phar] = df[col]
        else :
            temp_df[col] = df[col]
    #temp_df['store_code']=df['store_code']
    return temp_df

def getSalesDayPeriod(df):
    df['Morning'] = df['Morning']/df['total_sales']
    df['Afternoon'] = df['Afternoon']/df['total_sales']
    df['Evening'] = df['Evening']/df['total_sales']
    df['Night'] = df['Night']/df['total_sales']
    return df
    


sales_pdf_hr =  merge_hr(sales_pdf_daily)

sales_pdf_hr.head()
```

```python
sale_day_period_pdf = getSalesDayPeriod(sales_pdf_hr)
sale_day_period_pdf.head()
```
