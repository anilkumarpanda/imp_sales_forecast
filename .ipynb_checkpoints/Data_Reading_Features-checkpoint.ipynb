```python
# store_surrounding_pdf = pd.concat([sales_pdf, surrounding_pdf], axis=1, join='inner')
# store_surrounding_pdf.shape[0]
store_surrounding_pdf = surrounding_pdf[surrounding_pdf['store_code'].isin(sales_pdf['store_code'])]
store_surrounding_pdf.shape
```

```python
surrounding_amenities = []
store_surrounding_pdf.surroundings[0].keys()
```

```python
for i in range(store_surrounding_pdf.shape[0]):
    surrounding_amenities = store_surrounding_pdf.surroundings[0].keys()
    
store_surrounding_pdf_exp = pd.DataFrame(store_surrounding_pdf['store_code'])

for items in surrounding_amenities:
    #print(items)
    store_surrounding_pdf_exp[items] = store_surrounding_pdf.surroundings.apply(lambda x : x.get(items))
store_surrounding_pdf_exp.head()
```

```python
for items in surrounding_amenities:
    #print(items)
    store_surrounding_pdf_exp["no_of_"+items] = store_surrounding_pdf.surroundings.apply(lambda x : len(x.get(items)))
    #print(store_surrounding_pdf.surroundings.apply(lambda x : len(x.get(items))))
store_surrounding_pdf_exp.head()
```

```python
import numpy as np
def getAttributeList(dic,attribute):
    rating_list = []
    if len(dic)>0:
        for each in dic:
            for attr in attribute:
                rating_list.append(each.get(attr))
        rating_list = sum(filter(None,rating_list))/len(dic)
    return rating_list

attribute = ['rating']
getAttributeList(store_surrounding_pdf.surroundings[0].get('pharmacy'),attribute) #[2].get('rating')
for items in surrounding_amenities:
    #print(items)
    store_surrounding_pdf_exp["avg_rating_of_"+items] = store_surrounding_pdf.surroundings.apply(lambda x : getAttributeList(x.get(items),attribute))
    #print(store_surrounding_pdf.surroundings.apply(lambda x : getAttributeList(x.get(items),attribute)))
store_surrounding_pdf_exp.head()
```

```python
def getPostalCode(surrounding):
    surrounding_amenities = []
    surrounding_amenities = surrounding.keys()
    latitude=''
    longitude=''
    postcode= set()
    for items in surrounding_amenities:
        item_dict = surrounding.get(items)
        if len(item_dict)>0:
            for i in range(len(item_dict)):
                postcode.add(item_dict[i].get('address_components')[-1].get('short_name'))
    return list(postcode)
    
store_surrounding_pdf_exp['near_by_post_code'] = store_surrounding_pdf.surroundings.apply(lambda x : getPostalCode(x))
```

```python
store_post_pdf = store_surrounding_pdf_exp[['store_code','near_by_post_code']]
```

```python
store_post_pdf.head(20)
```

```python
from collections import defaultdict

postcode_store = defaultdict(list)

for index, row in store_post_pdf.iterrows():
    #print (row['store_code'], row['near_by_post_code'])
    for post_code in row['near_by_post_code']:
        postcode_store[post_code].append(row['store_code'])

postcode_store.get('2000')                              
```
