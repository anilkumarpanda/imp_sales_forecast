Creating aggregate information around each store, like no.of transit station around it etc

```python
import numpy as np
def getAttributeList(dic,attribute):
    rating_list = []
    if len(dic)>0:
        for each in dic:
            for attr in attribute:
                rating_list.append(each.get(attr))
        rating_list = sum(filter(None,rating_list))/len(dic)
    return rating_list

attribute = ['rating']
#getAttributeList(store_surrounding_pdf.surroundings[0].get('pharmacy'),attribute) #[2].get('rating')
for items in surrounding_amenities:
    #print(items)
    store_surrounding_pdf_exp["avr_"+items] = store_surrounding_pdf.surroundings.apply(lambda x : getAttributeList(x.get(items),attribute))
    #print(store_surrounding_pdf.surroundings.apply(lambda x : getAttributeList(x.get(items),attribute)))
store_surrounding_pdf_exp.head()
```

#### Presence of other stores, around a particular store may affect its sale.
Hence creating feature such as no.of store in the which share the same postal code, as the surrounding entity of each shop .

```python
def getCountofNearbyStores(post_code_list):
    global postcode_store
    temp_list = []
    result_list = []
    for post_code in post_code_list:
        for store in postcode_store.get(post_code):
            if store not in result_list:
                result_list.append(store)
    #print(result_list)
    return len(result_list)

store_surrounding_pdf_exp['no_of_near_by_shops'] = \
store_surrounding_pdf_exp.near_by_post_code.apply(lambda x : getCountofNearbyStores(x))
```

#### Sales may also be affected, by the city in which shop is located

```python
#Get City
def getCity(surrounding):
    surrounding_amenities = []
    surrounding_amenities = surrounding.keys()
    city = set()
    for items in surrounding_amenities:
        item_dict = surrounding.get(items)
        if len(item_dict)> 0:
            for i in range(len(item_dict)):
                if len(item_dict[i].get('address_components'))>=7:
                    #if len(item_dict[i].get('address_components')[-3].get('short_name'))==2:
                    city.add(item_dict[i].get('address_components')[-3].get('short_name'))
#                     else :
#                         city.add(item_dict[i].get('address_components')[-4].get('short_name'))
    return list(city)

store_surrounding_pdf_exp['city'] = store_surrounding_pdf.surroundings.apply(lambda x : getCity(x))
store_surrounding_pdf_exp['city']= store_surrounding_pdf_exp['city'].apply(lambda x : "Unknown" if len(x)==0 else x[0])
```

#### No.of review as a proxy for foot-fall within the locality of the store. 
#### Average review, as proxy for service likebility for each enitity .

```python
#Get No.of Review :
def getNReviews(surrounding):
    surrounding_amenities = []
    surrounding_amenities = surrounding.keys()
    n_review = []
    for items in surrounding_amenities:
        item_dict = surrounding.get(items)
        if len(item_dict)> 0:
            for i in range(len(item_dict)):
                n_review.append(item_dict[i].get('user_ratings_total'))
    return sum(filter(None,n_review))
store_surrounding_pdf_exp['n_review'] = store_surrounding_pdf.surroundings.apply(lambda x : getNReviews(x))
```

```python
entity_count_pdf = store_entity_pdf.groupby(['place_id']).count()
entity_count_pdf = entity_count_pdf.reset_index()
entity_count_pdf.head()

drop_list = ['store_id','key','total_user_rating','rating','post_code']
entity_count_pdf = entity_count_pdf.drop(drop_list,axis=1)
# store_entity_count_pdf.sort_values('place_id').head(20)
```

```python
#store_entity_pdf.join(entity_count_pdf,on='place_id',how='inner') 
store_entity_pdf = store_entity_pdf.merge(entity_count_pdf, left_on='place_id', right_on='place_id', how='inner')
store_entity_pdf = store_entity_pdf.drop_duplicates()
store_entity_pdf.sort_values('occurence_x').head(10)

store_entity_pdf = store_entity_pdf.drop('occurence_x',axis=1)
store_entity_pdf = store_entity_pdf.rename(columns={'occurence_y': 'nf_enity_store_id'}) 
```

```python
store_entity_pdf.sort_values('nf_enity_store_id',ascending=False).head(5)
```

```python
#store_entity_pdf = store_entity_pdf.drop(['inv_total_review','rating_per_review','enity_per_store'],axis=1)
store_entity_pdf['entity_per_store'] = 1/store_entity_pdf['nf_enity_store_id']
store_entity_pdf['n_rating_per_entity_store'] = store_entity_pdf['total_user_rating']/store_entity_pdf['entity_per_store']
store_entity_pdf['rating_per_entity_store'] = store_entity_pdf['rating'] / store_entity_pdf['entity_per_store']

store_entity_pdf.head(20)
```

```python
store_key_pdf = store_entity_pdf.groupby(['store_id','key']).mean()
store_key_pdf = store_key_pdf.reset_index()
store_key_pdf.head()
```
